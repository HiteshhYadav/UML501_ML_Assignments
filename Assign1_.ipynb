```python
#Ques 1
import numpy as np
arr = np.array([1, 2, 3, 6, 4, 5])
arr1 = np.flip(arr)
print(arr1)

arr2 = np.array([[1, 2, 3], [2, 4, 5], [1, 2, 3]])
arr2 = arr2.flatten()
print(arr2)

arr3 = np.array([[1, 2], [3, 4]])
arr4 = np.array([[1, 2], [3, 4]])
# print(arr3==arr4)
print(np.array_equal(arr3, arr4))

x = np.array([1, 2, 3, 4, 5, 1, 2, 1, 1, 1])
y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3])
ans1 = np.bincount(x).argmax()
indices = np.where(x == ans1)[0]
print("Element:", ans1)
print("Indices:", indices)
print()
ans2 = np.bincount(y).argmax()
indices2 = np.where(y == ans2)[0]
print("Element:", ans2)
print("Indices:", indices2)

gfg = np.matrix('[4,1,9;12,3,1;4,5,6]')
s = np.sum(gfg)
print("Total", s)
sr = np.sum(gfg, axis=1)
print(sr)
sc = np.sum(gfg, axis=0)
print(sc)

n_array = np.array([[55, 25, 15], [30, 44, 2], [11, 45, 77]])
dsum = np.trace(n_array)
print("Diagonal Sum:", dsum)

eig = np.linalg.eigvals(n_array)
print("Eigen", eig)

eigvalues, eigvectors = np.linalg.eig(n_array)
print(eigvectors)

n_inv = np.linalg.inv(n_array)
d = np.linalg.det(n_array)
print(n_inv)
print("Determinant: ", d)

p = np.array([[1, 2], [2, 3]])
q = np.array([[4, 5], [6, 7]])
r = np.dot(p, q)
print(r)

co = np.cov(p, q)
print("Covariance: ", co)

print()
p2 = np.array([[1, 2], [2, 3], [4, 5]])
q2 = np.array([[4, 5, 1], [6, 7, 2]])
r2 = np.dot(p2, q2)
print(r2)
co2 = np.cov(p2, q2.T)
print("Covariance: ", co2)

x = np.array([[2, 3, 4], [3, 2, 9]])
y = np.array([[1, 5, 0], [5, 10, 3]])
inner = np.inner(x, y)
outer = np.outer(x, y)
print(inner)
print()
print(outer)
print()

import itertools
cartesian = list(itertools.product(x, y))
print(cartesian)
```

```python
#Ques 4

import cv2

def img_to_array(path):
    # Read the image in color
    img = cv2.imread(path)

    if img is None:
        print("Error: Image not found or path is incorrect.")
        return

    # Check if image is grayscale
    if len(img.shape) == 2:
        # Grayscale image
        output_file = "grayscale_image_data.txt"
        np.savetxt(output_file, img, fmt='%d')
        print(f"Grayscale image saved to {output_file}")

    elif len(img.shape) == 3 and img.shape[2] == 3:
        # RGB image
        output_file = "rgb_image_data.txt"
        # Split into R, G, B channels
        r, g, b = cv2.split(img)
        with open(output_file, 'w') as f:
            f.write("Red Channel:\n")
            np.savetxt(f, r, fmt='%d')
            f.write("\nGreen Channel:\n")
            np.savetxt(f, g, fmt='%d')
            f.write("\nBlue Channel:\n")
            np.savetxt(f, b, fmt='%d')
        print(f"RGB image data saved to {output_file}")
    else:
        print("Unsupported image format.")

#call the function
img_to_array("/content/450-4503860_fourstar-services-consultations-workflow-transparent-hd-png-download.png")


#load it
import numpy as np

def load_rgb_image_from_txt(filepath):
    with open(filepath, 'r') as f:
        lines = f.readlines()

    # Split lines into RGB sections
    r_start = lines.index("Red Channel:\n") + 1
    g_start = lines.index("Green Channel:\n")
    b_start = lines.index("Blue Channel:\n")

    r_lines = lines[r_start:g_start - 1]
    g_lines = lines[g_start + 1:b_start - 1]
    b_lines = lines[b_start + 1:]

    r = np.array([[int(val) for val in line.split()] for line in r_lines], dtype=np.uint8)
    g = np.array([[int(val) for val in line.split()] for line in g_lines], dtype=np.uint8)
    b = np.array([[int(val) for val in line.split()] for line in b_lines], dtype=np.uint8)

    # Merge channels into RGB image
    rgb_img = np.stack((r, g, b), axis=-1)
    return rgb_img

# Load and display RGB image
rgb_img = load_rgb_image_from_txt("/content/rgb_image_data.txt")

import matplotlib.pyplot as plt
plt.imshow(cv2.cvtColor(rgb_img, cv2.COLOR_BGR2RGB))  # OpenCV loads in BGR
plt.title("RGB Image")
plt.axis('off')
plt.show()
```
